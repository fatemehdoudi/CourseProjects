## Part 1(a)
In this part, we'll train our generator and discriminator via the original minimax GAN objective:
<insert GAN Objective here>
$$min_{G} max_{D} \mathbb{E}_{x \sim p_{data}} [\log D(x)] + \mathbb{E}_{z \sim p(z)}[\log (1-D(G(z)))]$$

Use an MLP for both your generator and your discriminator, and train until the generated distribution resembles the target distribution.
* 3 layers
* 128 hidden dim
* LeakyReLU nonlinearities with negative_slope=0.2

### Solution

````python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

class DiscriminatorNet(nn.Module):
    def __init__(self):
        super(DiscriminatorNet, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(1, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.network(x)

class GeneratorNet(nn.Module):
    def __init__(self):
        super(GeneratorNet, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(1, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 1)
        )

    def forward(self, x):
        return self.network(x)

def noise(size):
    '''
    Generates a 1-d vector of gaussian sampled random values
    '''
    return torch.randn(size, 1)

def q1_a(train_data):
    N = train_data.shape[0]
    batch_size = 64
    discriminator = DiscriminatorNet()
    generator = GeneratorNet()
    optimizer = optim.Adam(list(discriminator.parameters()) + list(generator.parameters()), lr=0.001)
    loss = nn.BCELoss()
    d_losses = []
    samples_initial = []
    samples_final = []
    n_epochs = 100
    train_data = torch.tensor(train_data, dtype=torch.float32).view(-1, 1)

    for epoch in range(n_epochs):
        np.random.shuffle(train_data.numpy())
        d_loss_accum = 0
        num_batches = N // batch_size

        for i in range(num_batches):
            x = train_data[i * batch_size:(i + 1) * batch_size]
            fake_data = generator(noise(batch_size))

            one_labels = torch.ones(batch_size, 1)
            zero_labels = torch.zeros(batch_size, 1)

            optimizer.zero_grad()

            real_d = discriminator(x)
            fake_d = discriminator(fake_data.detach())

            real_loss = loss(real_d, one_labels)
            fake_loss = loss(fake_d, zero_labels)

            d_loss = real_loss + fake_loss
            d_loss_accum += d_loss.item()
            d_loss.backward()
            optimizer.step()

            if epoch == 0 and len(samples_initial) < 5000:
                samples_initial.extend(fake_data.detach().numpy().flatten())
            if epoch == n_epochs - 1 and len(samples_final) < 5000:
                samples_final.extend(fake_data.detach().numpy().flatten())

        d_losses.append(d_loss_accum / num_batches)

    samples_initial = np.array(samples_initial[:5000])
    samples_final = np.array(samples_final[:5000])

    x_linspace_initial = np.linspace(-1, 1, 1000)
    discriminator_output_initial = discriminator(torch.tensor(x_linspace_initial, dtype=torch.float32).view(-1, 1)).detach().numpy()

    x_linspace_final = np.linspace(-1, 1, 1000)
    discriminator_output_final = discriminator(torch.tensor(x_linspace_final, dtype=torch.float32).view(-1, 1)).detach().numpy()

    return np.array(d_losses), samples_initial, x_linspace_initial, discriminator_output_initial, samples_final, x_linspace_final, discriminator_output_final
````

## Solution

 ### Models

````python
class ResBlock(nn.Module):
    def __init__(self, in_dim, n_filters):
        super().__init__()
        self.conv1 = nn.Conv2d(in_dim, n_filters, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(n_filters, n_filters, kernel_size=3, stride=1, padding=1)
        self.batch_norm1 = nn.BatchNorm2d(n_filters)
        self.batch_norm2 = nn.BatchNorm2d(n_filters)
        self.shortcut = nn.Conv2d(in_dim, n_filters, kernel_size=1, stride=1, padding=0)

    def forward(self, x):
        shortcut = self.shortcut(x)
        x = self.batch_norm1(self.conv1(x))
        x = F.relu(x)
        x = self.batch_norm2(self.conv2(x))
        return F.relu(x + shortcut)

class DepthToSpace(nn.Module):
    def __init__(self, block_size):
        super().__init__()
        self.block_size = block_size
        self.block_size_sq = block_size * block_size

    def forward(self, input):
        output = input.permute(0, 2, 3, 1)
        (batch_size, d_height, d_width, d_depth) = output.size()
        s_depth = int(d_depth / self.block_size_sq)
        s_width = int(d_width * self.block_size)
        s_height = int(d_height * self.block_size)
        t_1 = output.reshape(batch_size, d_height, d_width, self.block_size_sq, s_depth)
        spl = t_1.split(self.block_size, 3)
        stack = [t_t.reshape(batch_size, d_height, s_width, s_depth) for t_t in spl]
        output = torch.stack(stack, 0).transpose(0, 1).permute(0, 2, 1, 3, 4).reshape(batch_size, s_height, s_width, s_depth)
        output = output.permute(0, 3, 1, 2)
        return output

# class SpaceToDepth(nn.Module):
#     def __init__(self, block_size):
#         super().__init__()
#         self.block_size = block_size
#         self.block_size_sq = block_size * block_size

#     def forward(self, input):
#         output = input.permute(0, 2, 3, 1)
#         (batch_size, s_height, s_width, s_depth) = output.size()
#         d_depth = s_depth * self.block_size_sq
#         d_width = int(s_width / self.block_size)
#         d_height = int(s_height / self.block_size)
#         t_1 = output.split(self.block_size, 2)
#         stack = [t_t.reshape(batch_size, d_height, d_depth) for t_t in t_1]
#         output = torch.stack(stack, 1)
#         output = output.permute(0, 2, 1, 3)
#         output = output.permute(0, 3, 1, 2)
#         return output
class SpaceToDepth(nn.Module):
    def __init__(self, block_size):
        super().__init__()
        self.block_size = block_size
        self.block_size_sq = block_size * block_size

    def forward(self, input):
        output = input.permute(0, 2, 3, 1)
        (batch_size, s_height, s_width, s_depth) = output.size()
        d_depth = s_depth * self.block_size_sq
        d_width = int(s_width / self.block_size)
        d_height = int(s_height / self.block_size)
        t_1 = output.split(self.block_size, 2)
        stack = [t_t.reshape(batch_size, d_height, d_depth) for t_t in t_1]
        output = torch.stack(stack, 1)
        output = output.permute(0, 2, 1, 3)
        output = output.permute(0, 3, 1, 2)
        return output

class Upsample_Conv2d(nn.Module):
  def __init__(self, in_dim, out_dim, kernel_size=(3, 3), stride=1, padding=1):
    super().__init__()
    self.in_dim = in_dim
    self.out_dim = out_dim
    self.kernel_size = kernel_size
    self.stride = stride
    self.padding = padding
    self.conv = nn.Conv2d(in_dim, out_dim, kernel_size, stride=stride, padding=padding)
    self.depth_to_block = DepthToSpace(block_size = 2)

  def forward(self, x):
    x = torch.cat([x, x, x, x], dim=1)
    x = self.depth_to_block(x)
    x = self.conv(x)
    return x

class Downsample_Conv2d(nn.Module):
  def __init__(self, in_dim, out_dim, kernel_size=(3, 3), stride=1, padding=1):
    super().__init__()
    self.in_dim = in_dim
    self.out_dim = out_dim
    self.kernel_size = kernel_size
    self.stride = stride
    self.padding = padding
    self.conv = nn.Conv2d(in_dim, out_dim, kernel_size, stride=stride, padding=padding)
    self.space_to_depth = SpaceToDepth(block_size = 2)

  def forward(self, x):
    x = self.space_to_depth(x)
    x = torch.stack(x.chunk(4, dim=1), dim=0).sum(dim=0) / 4.0
    x = self.conv(x)
    return x

class ResnetBlockUp(nn.Module):
  def __init__(self, in_dim, kernel_size=(3, 3), n_filters=256):
    super().__init__()
    self.in_dim = in_dim
    self.kernel_size = kernel_size
    self.n_filters = n_filters
    self.network = nn.Sequential(nn.BatchNorm2d(in_dim),
                                 nn.ReLU(),
                                 nn.Conv2d(in_dim, n_filters, kernel_size, padding=1),
                                 nn.BatchNorm2d(n_filters),
                                 nn.ReLU()
                                 )
    self.residual = Upsample_Conv2d(n_filters, n_filters, kernel_size=kernel_size, padding=1)
    self.shortcut = Upsample_Conv2d(in_dim, n_filters, kernel_size=(1, 1), padding=0)

  def forward(self, x):
    shortcut = self.shortcut(x)
    x = self.network(x)
    residual = self.residual(x)
    return residual + shortcut

class ResnetBlockDown(nn.Module):
  def __init__(self, in_dim, kernel_size=(3, 3), n_filters=256):
    super().__init__()
    self.in_dim = in_dim
    self.kernel_size = kernel_size
    self.n_filters = n_filters
    self.network = nn.Sequential(nn.ReLU(),
                                 nn.Conv2d(in_dim, n_filters, kernel_size, padding=1),
                                 nn.ReLU()
                                 )
    self.residual = Downsample_Conv2d(n_filters, n_filters, kernel_size=kernel_size, padding=1)
    self.shortcut = Downsample_Conv2d(in_dim, n_filters, kernel_size=(1, 1), padding=0)

  def forward(self, x):
    _x = x.clone()
    _x = self.network(_x)
    residual = self.residual(_x)
    shortcut = self.shortcut(x)
    return residual + shortcut


class Generator(nn.Module):
  def __init__(self, n_samples=1024, n_filters=128):
    super().__init__()
    self.n_samples = n_samples
    self.n_filters = n_filters
    self.linear = nn.Linear(128, 4*4*256)
    self.network = nn.Sequential(ResnetBlockUp(in_dim=256, n_filters=n_filters),
                                 ResnetBlockUp(in_dim=n_filters, n_filters=n_filters),
                                 ResnetBlockUp(in_dim=n_filters, n_filters=n_filters),
                                 nn.BatchNorm2d(n_filters),
                                 nn.ReLU(),
                                 nn.Conv2d(n_filters, 3, kernel_size=(3, 3), padding=1),
                                 nn.Tanh())

  def forward(self, x):
    x = self.linear(x)
    x = x.view(-1, 256, 4, 4)
    x = self.network(x)
    return x


class Discriminator(nn.Module):
  def __init__(self, n_filters=128):
    super().__init__()
    self.n_filters = n_filters
    self.network = nn.Sequential(ResnetBlockDown(3, n_filters=n_filters),
                                 ResnetBlockDown(128, n_filters=n_filters),
                                 ResBlock(n_filters, n_filters=n_filters),
                                 ResBlock(n_filters, n_filters=n_filters),
                                 nn.ReLU())
    self.sum_pooling = nn.AdaptiveAvgPool2d((1, 1))
    self.linear = nn.Linear(128, 1)
  def forward(self, x):
      x = self.network(x)
      x = self.sum_pooling(x)
      x = x.view(-1, 128)
      x = self.linear(x)
      return x

import torch.autograd as autograd


def q2(train_data):
    """
    train_data: An (n_train, 3, 32, 32) numpy array of CIFAR-10 images with values in [0, 1]

    Returns
    - a (# of training iterations,) numpy array of WGAN critic train losses evaluated every minibatch
    - a (1000, 32, 32, 3) numpy array of samples from your model in [0, 1].
        The first 100 will be displayed, and the rest will be used to calculate the Inception score.
    """

    """ YOUR CODE HERE """
    N = train_data.shape[0]
    n_filters=128
    train_data = torch.tensor(train_data, dtype=torch.float32)
    n_epochs = 20
    batch_size = 256
    lambda_ = 10
    n_critics = 5

    generator = Generator(n_filters=n_filters)
    discriminator = Discriminator(n_filters=n_filters)

    generator_optimizer = optim.Adam(generator.parameters(), lr=2e-4, betas=(0 ,0.9))
    discriminator_optimizer = optim.Adam(discriminator.parameters(), lr=2e-4, betas=(0 ,0.9))

    critic_losses = []
    samples = []

    for epoch in range(n_epochs):
      np.random.shuffle(train_data.numpy())
      loss_accum = 0
      num_batches = N//batch_size
      for i in range(num_batches):
        real_data = train_data[i*batch_size:(i+1)*batch_size , : , :, :]
        for i in range(n_critics):
          eps = torch.randn(1)
          noise_ = torch.randn(batch_size , 128)
          fake_data = generator(noise_)

          x_hat = eps*real_data + (1-eps)*fake_data

          D_x_hat = discriminator(x_hat)
          ones_ = torch.ones_like(D_x_hat)
          gradients = autograd.grad(
            outputs=D_x_hat,
            inputs=x_hat,
            grad_outputs=ones_,
            create_graph=True,
            retain_graph=True,
          )[0]
          # gradients = gradients.view(batch_size, -1)
          gradients = gradients.reshape(batch_size, -1)

          grad_norm = gradients.norm(2, 1)

          discriminator_optimizer.zero_grad()
          loss_D = -torch.mean(discriminator(real_data)) + torch.mean(discriminator(fake_data)) + lambda_ * torch.mean((grad_norm - 1) ** 2)
          loss_D.backward()
          discriminator_optimizer.step()

          for p in discriminator.parameters():
              p.data.clamp_(-0.01, 0.01)

        critic_losses.append(loss_D.item())
        #training Generator
        generator_optimizer.zero_grad()
        noise_ = torch.randn(batch_size, 128)
        fake_data = generator(noise_)
        loss_G = -torch.mean(discriminator(fake_data))
        loss_G.backward()
        generator_optimizer.step
        if epoch % 100 == 0:
            print(f'Epoch [{epoch}/{n_epochs}] - Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}')

    with torch.no_grad():
      noise_ = torch.randn(1000, 128).to('cuda')
      samples = generator(noise_).cpu().numpy()
      samples = (samples + 1) / 2
      samples = samples.transpose(0, 2, 3, 1)

    return critic_losses, samples
````

In [CycleGAN](https://arxiv.org/pdf/1703.10593.pdf), the goal is to learn functions $F$ and $G$ that can transform images from $X \rightarrow Y$ and vice-versa. This is an unconstrained problem, so we additionally enforce the *cycle-consistency* property, where we want
$$x \approx G(F(x))$$
and  
$$y \approx F(G(x))$$
This loss function encourages $F$ and $G$ to approximately invert each other. In addition to this cycle-consistency loss, we also have a standard GAN loss such that $F(x)$ and $G(y)$ look like real images from the other domain.
